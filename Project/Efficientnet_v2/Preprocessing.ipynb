```python
import os
import urllib.request
import tarfile
```

```python
# URL of the tar file
for i in range(1, 6):

    if i <= 9:
        padded_number = str(i).zfill(2)
    else:
        padded_number = str(i)

    filename = f"fish_{padded_number}.tar"
    print(filename)
    url = f"https://groups.inf.ed.ac.uk/vision/DATASETS/FISH4KNOWLEDGE/WEBSITE/GROUNDTRUTH/RECOG/Archive/fish_image/{filename}"
    # Download the tar file
    tar_filename = os.path.join("raw/images", filename)
    urllib.request.urlretrieve(url, tar_filename)

    # Extract the contents of the tar file
    with tarfile.open(tar_filename, 'r') as tar:
        tar.extractall(path="raw/images")
 
    # Clean up: remove the downloaded tar file
    os.remove(tar_filename)

print("Download and extraction complete.")
```

```python
import os
import shutil
import random

dest_path = "data"

# Limit the loop to the first 5000 files
file_limit = 2500

for i in range(1, 6):

    if i <= 9:
        padded_number = str(i).zfill(2)
    else:
        padded_number = str(i)

    print(padded_number)
    
    # Counter for processed files
    file_count = 0

    image_path = f"raw/images/fish_{padded_number}"

    files = [f for f in os.listdir(image_path) if f.endswith(('.png', '.txt'))]

    # Randomly select file_limit number of files
    selected_files = random.sample(files, min(file_limit, len(files)))
    
    for image_filename in selected_files:
        full_image_path = os.path.join(image_path, image_filename)
        
        image_destination_path = f"raw/prepared/{padded_number}"
        os.makedirs(image_destination_path, exist_ok=True)
        
        image_destination_filename = f"{i}_{image_filename}"
        destination_path = os.path.join(image_destination_path, image_destination_filename)
        shutil.copy(full_image_path, destination_path)
        
        file_count += 1
```

```python
import os

def remove_files_in_directory(directory):
    for file_name in os.listdir(directory):
        file_path = os.path.join(directory, file_name)
        try:
            if os.path.isfile(file_path):
                os.remove(file_path)
                #print(f"Removed: {file_path}")
        except Exception as e:
            print(f"Error while deleting {file_path}: {e}")

for i in range(1, 6):

    if i <= 9:
        padded_number = str(i).zfill(2)
    else:
        padded_number = str(i)
    
    remove_files_in_directory(f'raw/prepared/{padded_number}')
    remove_files_in_directory(f'data/train/{padded_number}')
    remove_files_in_directory(f'data/validate/{padded_number}')
    remove_files_in_directory(f'data/test/{padded_number}')
```

```python
import os
import random
import shutil

# Set the path to your dataset folder
dataset_path = "raw"

# Set the ratio of data to be used for validation
validation_ratio = 0.2  # Adjust this according to your needs
test_ratio = 0.1

# Create directories for the train and validation sets
train_path = "data/train/"
val_path = "data/validate/"
test_path = "data/test/"

os.makedirs(train_path, exist_ok=True)
os.makedirs(val_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# Get the list of images in the "images" folder
image_folder = os.path.join(dataset_path, "prepared")

for i in range(1, 6):

    if i <= 9:
        padded_number = str(i).zfill(2)
    else:
        padded_number = str(i)
    
    all_images = [f for f in os.listdir(f"{image_folder}/{padded_number}") if f.endswith('.png') and f.startswith(f"{str(i)}_")]
    
    # Calculate the number of images to be used for validation
    num_val_images = int(validation_ratio * len(all_images))

    # Calculate the number of images to be used for test
    num_test_images = int(test_ratio * len(all_images))

    # Randomly select images for validation
    val_images = random.sample(all_images, num_val_images)

    # Randomly select images for test
    test_images = random.sample(all_images, num_test_images)

    # Move images and labels to the appropriate directories
    for image_name in all_images:
        image_path = os.path.join(f"{image_folder}/{padded_number}", image_name)
        os.makedirs(os.path.dirname(image_path), exist_ok=True)
        if image_name in val_images:
            shutil.copy(image_path, f"{val_path}/{padded_number}")
        elif image_name in test_images:
            shutil.copy(image_path, f"{test_path}/{padded_number}")
        else:
            shutil.copy(image_path, f"{train_path}/{padded_number}")
```

```python
import shutil

def zip_directory(directory_path, zip_file_path):
    """
    Zip a directory.

    Parameters:
    - directory_path (str): The path of the directory to be zipped.
    - zip_file_path (str): The path of the zip file to be created.
    """
    shutil.make_archive(zip_file_path, 'zip', directory_path)

# Example usage:
directory_to_zip = 'data/test'
zip_file_path = 'dataset.zip'

zip_directory(directory_to_zip, zip_file_path)
```
